```python
import pandas as ps
```

```python
tesla_df = ps.read_csv('TSLA.csv')
tesla_df.tail()
```

```python
tesla_df = tesla_df.loc[tesla_df['Close'] // 6 != 0]
tesla_df
```

```python
tesla_df['Close'].describe()
```

```python
filter = tesla_df.notna()
```

```python
tesla_df = tesla_df[filter]
tesla_df
```

```python
g = sns.pairplot(
    tesla_df,
    x_vars = ['Date'],
    y_vars = ['Close', ],
    hue='Date'
)
g.fig.set_size_inches(15, 15) 
```

```python
open_max = tesla_df['Open'].max()
open_min = tesla_df['Open'].min()


def normal_date(one_serias):
    return (one_serias - open_min) / (open_max - open_min)

tesla_df['Open'] = tesla_df['Open'].apply(normal_date)
tesla_df
```

```python


new_df = ps.DataFrame(columns = ['Date', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume'])

for i in range(0, 10):
    new_vector = []
    for j in range (0, 7):
        new_vector.append((i+j) ** j)
        
    new_df.loc[len(new_df.index)] = new_vector
```

```python
tesla_df = ps.concat([tesla_df, new_df], ignore_index=True)
tesla_df.tail(10)
```
