#### Метод Эйлера для ДУ первого порядка

```python
import pandas as pd
import numpy as np
from math import tan
```

```python
table_res_eler =  pd.DataFrame(columns = ['xi',
                                    'yi',
                                    'f(xi, yi)',
                                    'h * f(xi, yi)'])
```

```python
class StartParametrs():
    start_x: float #x0 начальное положение
    start_y: float #у0 начальное положение
    n_iterval: int #нижняя граница интервала
    e_interval: int #верхняя граница интервала
    h: int #кол-во шагов
    size_iter: float
    
    status = True #Флаг сигнализирующий, что мы достигли верхней границы интервала 

        
    def __init__(self, start_x, start_y, n_interval, e_interval, h):
        """ start_x - , start_y, n_interval, e_interval, h  """
        self.start_x = start_x
        self.start_y = start_y
        self.n_interval = n_interval
        self.e_interval = e_interval
        self.h = h
        self.size_iter = (e_interval - n_interval)/h
        
    def next_step(self): #Создание нового шага
        self.n_interval += self.size_iter
        
        if self.n_interval > self.e_interval:
            self.status = False
```

```python
def_data = StartParametrs(0, 2, 0, 2, 10) #Создание объкта для метода эйлера
```

```python
def info_add_table(table, new_data):
    table.loc[len(table.index)] = new_data
```

```python
def function(x, y):
    return tan((x ** 2 + y ** 2) / (1 + x **2 + y ** 2)) #Заданная функция
```

```python
def next_iter(new_step, cut_status, table_name):
    res = function(new_step[0], new_step[1])
    new_step = np.append(new_step, res)
    new_step = np.append(new_step, res * def_data.size_iter)
    
    if cut_status:
        new_step = cut_coef(new_step)
    
    info_add_table(table_name, new_step)
    
    return new_step
```

```python
one_iter = np.array([def_data.n_interval, def_data.start_y])

while def_data.status:    
    one_iter = next_iter(one_iter, False, table_res_eler)
    def_data.next_step()
    
    
    one_iter = np.array([def_data.n_interval, one_iter[1] + one_iter[3]])    
```

```python
table_res_cut =  pd.DataFrame(columns = ['xi',
                                         'yi_cut',
                                         'f(xi, yi)',
                                         'h * f(xi, yi)',
                                         'K1',
                                         'K2',
                                         'K3',
                                         'K4'])
```

```python
cut_m = StartParametrs(0, 2, 0, 2, 10) #Создание объкта для метода Рунге-Кутта
```

```python
def cut_coef(new_iter):
    step = cut_m.size_iter
    
    k1 = new_iter[-1]
    k2 = function(new_iter[0] + step/2, new_iter[1] + step/2*k1)
    k3 = function(new_iter[0] + step/2, new_iter[1] + step/2*k2)
    k4 = function(new_iter[0] + step, new_iter[1] + step*k1)

    new_iter = np.append(new_iter, [k1, k2, k3, k4])
    #print(new_iter)
    return new_iter
```

```python
def cut_yi(new_iter):
    return new_iter[1] + (cut_m.size_iter * (new_iter[-4] + 2 * new_iter[-3] + 2 * new_iter[-2] + new_iter[-1]))/6
```

```python
one_iter = np.array([cut_m.n_interval, cut_m.start_y])

while cut_m.status:    
    one_iter = next_iter(one_iter, True, table_res_cut)
    cut_m.next_step()
    
    one_iter = np.array([cut_m.n_interval, cut_yi(one_iter)])  
```
